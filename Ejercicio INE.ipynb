DATOS DE LA INE:

-NOMBRE
-DOMICILIO
    -Calle
    -Manzana
    -Lote
    -SuperManzana
    -Municipio
    -Estado
    -Código Postal
-CLAVE ELECTORAL
-CURP
-FECHA DE NACIMIENTO
-AÑO DE REGISTRO
-SECCIÓN
-VIGENCIA
-SEXO

```python
def ask_data():
    
    print("Ingrese los siguientes datos de su INE por favor (✿◡‿◡)")
    
    ine_data = {}
    ine_data ['nombre'] = input("Su nombre es:")
    ine_data ['calle'] = input("Su número de calle es:")
    ine_data ['manzana'] = input("Su número de manzana es: ")
    ine_data ['lote'] = input("Su número de lote es: ")
    ine_data ['super_manzana'] = input("Su super manzana es: ")
    ine_data ['municipio'] = input("Su municipio es: ")
    ine_data ['estado'] = input("Su estado es: ")
    ine_data ['codigo_postal'] = input("Su código postal es: ")
    ine_data ['clave_elector'] = input("Su clave de elector es: ")
    ine_data ['curp'] = input("Su CURP es: ")
    ine_data ['fecha_nacimiento'] = input("Su fecha de nacimiento es: ")
    ine_data ['registro'] = input("Su año de registro es: ")
    ine_data ['seccion'] = input("Su sección es: ")
    ine_data ['vigencia'] = input("La fecha de vigencia de su INE es: ")
    ine_data ['sexo'] = input("¿Usted es biológicamente mujer u hombre? Ponga M si es mujer y H si es hombre: ")
    return ine_data

data = ask_data()

def save_ine_data(data):
    with open('INE_Data.txt', 'w') as saved_data:
        saved_data.write("Sus datos son:\n")
        saved_data.write("\n")
        saved_data.write(f"   Nombre completo: {data['nombre']}\n")
        saved_data.write("Su domicilio es:\n")
        saved_data.write(f"   Calle: {data['calle']}\n")
        saved_data.write(f"   Manzana: {data['manzana']}\n")
        saved_data.write(f"   Lote: {data['lote']}\n")
        saved_data.write(f"   Super Manzana: {data['super_manzana']}\n")
        saved_data.write(f"   Municipio: {data['municipio']}\n")
        saved_data.write(f"   Estado: {data['estado']}\n")
        saved_data.write(f"   Codigo Postal: {data['codigo_postal']}\n")
        saved_data.write("Otros datos:\n")
        saved_data.write(f"   Clave de elector: {data['clave_elector']}\n")
        saved_data.write(f"   CURP: {data['curp']}\n")
        saved_data.write(f"   Fecha de nacimiento: {data['fecha_nacimiento']}\n")
        saved_data.write(f"   Registro: {data['registro']}\n")
        saved_data.write(f"   Seccion: {data['seccion']}\n")
        saved_data.write(f"   Vigencia: {data['vigencia']}\n")
        saved_data.write(f"   Sexo: {data['sexo']}\n")

save_ine_data(data)

print("\n")
print("Muchas gracias por proveernos sus datos, estos fueron guardados en INE_Data.txt\n")
print("Estos son los datos que usted guardó en dicho documento: \n")

with open("INE_Data.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)
```
